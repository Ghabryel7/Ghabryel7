```python
import numpy as np
```

```python
url= 'https://raw.githubusercontent.com/allanspadini/numpy/dados/citrus.csv'
```

```python
dado = np.loadtxt(url, delimiter=(','), usecols=np.arange(1,6,1), skiprows=1)
```

```python
dado
```

```python
diam_orange = dado[:5000,0]
diam_grape = dado[5000:,0]
peso_orange = dado[:5000,1]
peso_grape = dado[5000:,1]
```

```python
import matplotlib.pyplot as plt

plt.plot(diam_orange, peso_orange)
plt.plot(diam_grape, peso_grape)
```

```python
Y = peso_orange
X = diam_orange
n = np.size(X)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - (np.sum(X))**2)
b = b = np.mean(Y) - a*np.mean(X)
y = y = a*X+b
```

```python
plt.plot(diam_orange, peso_orange)

plt.plot(X, y)
```

```python
Y = peso_grape
X = diam_grape
n = np.size(X)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - (np.sum(X))**2)
b = b = np.mean(Y) - a*np.mean(X)
y = y = a*X+b
b
```

```python
plt.plot(diam_grape, peso_grape)
plt.plot(X, y)
```

```python
norma = np.array([])
np.random.seed(84)
coef_angulares = np.random.uniform(low=0.0,high=30.0,size=100)

for i in range(100):
  norma = np.append(norma,np.linalg.norm(Y- (coef_angulares[i]*X+b)))
```
